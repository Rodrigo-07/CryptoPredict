```python
import yfinance as yf
import pandas as pd
```

Vou fazer uma analise de duas moedas:

- Cardano (ADA-USD)
- Solana (SOL-USD)



```python
# Definir começo e fim

start_ada = '2020-05-01'
end_ada = '2023-09-24'

start_sol = '2017-05-01'
end_sol = '2023-09-24'
```

```python
# Dados do ativo

ada_data = yf.download('ADA-USD', start=start_ada, end=end_ada)

sol_data = yf.download('SOL-USD', start=start_sol, end=end_sol)
```

```python
ada_data
```

```python
sol_data
```

# Analise

```python
ada_data.describe()
```

Podemos ver que o Cardano tem um preço baixo em toda a sua historia e que teve apenas uma subida.

```python
sol_data.describe()
```

No caso da Solana vemos um cresscimento grande indo de 0.5 para 200.

```python
import matplotlib.pyplot as plt

# Preço de fechamento do Cardano
ada_data['Close'].plot(title='Fechamento preços Cardano')
plt.show()

# Preço de fechamento do Solana
sol_data['Close'].plot(title='Fechamento preços Solana')
plt.show()
```

```python
# Gráficos de média móvel de 20 dias

ada_data['Close'].plot(title='Fechamento preços Cardano')
ada_data['Close'].rolling(window=20).mean().plot(label='Média Móvel 20 dias')
plt.show()

sol_data['Close'].plot(title='Fechamento preços Solana')
sol_data['Close'].rolling(window=20).mean().plot(label='Média Móvel 20 dias')
plt.show()
```

# Teste com regressão linear

```python
X = sol_data[['Open', 'High', 'Low']]
Y = sol_data['Close']
```

```python
# Trainar o modelo de regressão linear

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, Y)
```

```python
# Pegar dados a partir end_date

sol_data_real = yf.download('SOL-USD', start=end_ada, end=None)


```

```python
sol_data_real
```

```python
# Prever o preço de fechamento

sol_data_real['ClosePredict'] = model.predict(sol_data_real[['Open', 'High', 'Low']])
```

```python
sol_data_real['Close'].plot(title='Previsão de preço de fechamento do Cardano')
sol_data_real['ClosePredict'].plot(label='Previsão')
plt.show()
```
