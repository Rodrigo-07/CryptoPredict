```python
from sklearn.preprocessing import MinMaxScaler
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense

class CryptoPredicter:
    def __init__(self, CryptoName, start_date, end_date, prediction_days=60):
        self.CryptoName = CryptoName
        self.start_date = start_date
        self.end_date = end_date
        self.prediction_days = prediction_days
        
    def get_data(self):
        data = yf.download(self.CryptoName, start=self.start_date, end=self.end_date)
        print(data)
        return data

    def train_model(self, data):
        close_data = data['Close'].values.reshape(-1, 1)
        scaler = MinMaxScaler(feature_range=(0, 1))  # Normalização no intervalo 0-1
        scaled_close_data = scaler.fit_transform(close_data)

        # Separando treino e teste
        train_size = int(len(scaled_close_data) * 0.8)
        train_data = scaled_close_data[:train_size]
        test_data = scaled_close_data[train_size - self.prediction_days:]

        # Preparando dados de treino
        x_train, y_train = [], []
        for x in range(self.prediction_days, len(train_data)):
            x_train.append(train_data[x-self.prediction_days:x, 0])
            y_train.append(train_data[x, 0])

        x_train, y_train = np.array(x_train), np.array(y_train)
        x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

        # Definição do modelo com Dropout para evitar overfitting
        model = Sequential()
        model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
        model.add(Dropout(0.2))  # Dropout adicionado
        model.add(LSTM(units=50))
        model.add(Dropout(0.2))
        model.add(Dense(units=1))

        # Compilação e treinamento
        model.compile(optimizer='adam', loss='mean_squared_error')
        model.fit(x_train, y_train, epochs=25, batch_size=32)

        model.save(f'{self.CryptoName}_model.h5')
        return model, scaler
    
    def predict(self, model, data, scaler, days_to_predict=30):
        close_data = data['Close'].values.reshape(-1, 1)
        scaled_close_data = scaler.transform(close_data)

        # Preparando dados de teste
        x_test = []
        for x in range(self.prediction_days, len(scaled_close_data)):
            x_test.append(scaled_close_data[x-self.prediction_days:x, 0])

        x_test = np.array(x_test)
        x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

        predicted_prices = model.predict(x_test)
        predicted_prices = scaler.inverse_transform(predicted_prices)
        return predicted_prices

    def forecast(self, model, data, scaler, days_to_predict=30):
        close_data = data['Close'].values.reshape(-1, 1)
        scaled_close_data = scaler.transform(close_data)
        last_prediction_days = scaled_close_data[-self.prediction_days:]

        predicted_prices = []
        x_input = last_prediction_days.reshape(1, self.prediction_days, 1)

        for _ in range(days_to_predict):
            prediction = model.predict(x_input, verbose=0)
            predicted_prices.append(prediction[0][0])
            prediction_reshaped = prediction.reshape(1, 1, 1)
            x_input = np.concatenate((x_input[:, 1:, :], prediction_reshaped), axis=1)

        predicted_prices_array = np.array(predicted_prices).reshape(-1, 1)
        predicted_prices_actual = scaler.inverse_transform(predicted_prices_array)

        last_date = data.index[-1]
        future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=days_to_predict, freq='D')

        forecast_df = pd.DataFrame({'Date': future_dates, 'Predicted_Close': predicted_prices_actual.flatten()})
        forecast_df.set_index('Date', inplace=True)
        return forecast_df
    
    def plot_forecast(self, forecast_df, real_data):
        plt.figure(figsize=(14, 7))
        plt.plot(real_data.index, real_data['Close'], label='Preço Real')
        plt.plot(forecast_df.index, forecast_df['Predicted_Close'], label='Previsão', linestyle='--')
        plt.title(f'Previsão dos Preços de Fechamento da {self.CryptoName} para os Próximos 30 Dias')
        plt.xlabel('Data')
        plt.ylabel('Preço (USD)')
        plt.legend()
        plt.grid(True)
        plt.show()

# Uso do modelo
SolanaPredicter = CryptoPredicter('SOL-USD', '2020-01-01', '2024-09-01')

sol_data = SolanaPredicter.get_data()

model, scaler = SolanaPredicter.train_model(sol_data)

predicted_prices = SolanaPredicter.forecast(model, sol_data, scaler)

SolanaPredicter.plot_forecast(predicted_prices, sol_data)

predicted_prices
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

class CryptoPredicter:
    def __init__(self, CryptoName, start_date, end_date, prediction_days=60):
        self.CryptoName = CryptoName
        self.start_date = start_date
        self.end_date = end_date
        self.prediction_days = prediction_days
        
    def get_data(self):
        data = yf.download(self.CryptoName, start=self.start_date, end=self.end_date)
        print(data)
        return data

    def train_model(self, data):
        close_data = data['Close'].values.reshape(-1, 1)
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_close_data = scaler.fit_transform(close_data)

        # Separando treino e teste
        train_size = int(len(scaled_close_data) * 0.8)
        train_data = scaled_close_data[:train_size]
        test_data = scaled_close_data[train_size - self.prediction_days:]

        # Preparando dados de treino
        x_train, y_train = [], []
        for x in range(self.prediction_days, len(train_data)):
            x_train.append(train_data[x-self.prediction_days:x, 0])
            y_train.append(train_data[x, 0])

        x_train, y_train = np.array(x_train), np.array(y_train)
        x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

        # Definição do modelo
        model = Sequential()
        model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
        model.add(Dropout(0.2))
        model.add(LSTM(units=50, return_sequences=True))
        model.add(Dropout(0.2))
        model.add(LSTM(units=50))
        model.add(Dropout(0.2))
        model.add(Dense(units=1))  # Prediction of the next closing value

        # Compilação e treinamento
        model.compile(optimizer='adam', loss='mean_squared_error')
        model.fit(x_train, y_train, epochs=25, batch_size=32)

        model.save(f'{self.CryptoName}_model.h5')
        return model, scaler, test_data
    
    def predict(self, model, data, scaler):
        close_data = data['Close'].values.reshape(-1, 1)
        scaled_close_data = scaler.transform(close_data)

        # Preparando dados de teste
        x_test = []
        for x in range(self.prediction_days, len(scaled_close_data)):
            x_test.append(scaled_close_data[x-self.prediction_days:x, 0])

        x_test = np.array(x_test)
        x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

        predicted_prices = model.predict(x_test)
        predicted_prices = scaler.inverse_transform(predicted_prices)
        return predicted_prices
    
    def forecast(self, model, data, scaler, days_to_predict=30):
        close_data = data['Close'].values.reshape(-1, 1)
        scaled_close_data = scaler.transform(close_data)
        last_prediction_days = scaled_close_data[-self.prediction_days:]

        predicted_prices = []
        x_input = last_prediction_days.reshape(1, self.prediction_days, 1)

        for _ in range(days_to_predict):
            prediction = model.predict(x_input, verbose=0)
            predicted_prices.append(prediction[0][0])
            prediction_reshaped = prediction.reshape(1, 1, 1)
            x_input = np.concatenate((x_input[:, 1:, :], prediction_reshaped), axis=1)

        predicted_prices_array = np.array(predicted_prices).reshape(-1, 1)
        predicted_prices_actual = scaler.inverse_transform(predicted_prices_array)

        last_date = data.index[-1]
        future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=days_to_predict, freq='D')

        forecast_df = pd.DataFrame({'Date': future_dates, 'Predicted_Close': predicted_prices_actual.flatten()})
        forecast_df.set_index('Date', inplace=True)
        return forecast_df

    def evaluate_model(self, actual, predicted):
        # Garantir que os tamanhos dos dados sejam iguais
        if len(actual) != len(predicted):
            actual = actual[-len(predicted):]

        mae = mean_absolute_error(actual, predicted)
        mse = mean_squared_error(actual, predicted)
        rmse = np.sqrt(mse)
        mape = np.mean(np.abs((actual - predicted) / actual)) * 100
        
        print(f"Mean Absolute Error (MAE): {mae}")
        print(f"Mean Squared Error (MSE): {mse}")
        print(f"Root Mean Squared Error (RMSE): {rmse}")
        print(f"Mean Absolute Percentage Error (MAPE): {mape}%")

        return mae, mse, rmse, mape
    
    def plot_forecast(self, forecast_df, real_data):
        plt.figure(figsize=(14, 7))
        plt.plot(real_data.index, real_data['Close'], label='Preço Real')
        plt.plot(forecast_df.index, forecast_df['Predicted_Close'], label='Previsão', linestyle='--')
        plt.title(f'Previsão dos Preços de Fechamento da {self.CryptoName} para os Próximos 30 Dias')
        plt.xlabel('Data')
        plt.ylabel('Preço (USD)')
        plt.legend()
        plt.grid(True)
        plt.show()

# Uso do modelo
SolanaPredicter = CryptoPredicter('SOL-USD', '2020-01-01', '2024-09-01')

sol_data = SolanaPredicter.get_data()

model, scaler, test_data = SolanaPredicter.train_model(sol_data)

predicted_prices_df = SolanaPredicter.forecast(model, sol_data, scaler)

# Validando as previsões
predicted_prices = predicted_prices_df['Predicted_Close'].values
actual_prices = sol_data['Close'].values[-len(predicted_prices):]  # Ajusta o tamanho para corresponder às previsões
SolanaPredicter.evaluate_model(actual_prices, predicted_prices)

SolanaPredicter.plot_forecast(predicted_prices_df, sol_data)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split

class CryptoPredicterWithBoosting:
    def __init__(self, CryptoName, start_date, end_date, prediction_days=60):
        self.CryptoName = CryptoName
        self.start_date = start_date
        self.end_date = end_date
        self.prediction_days = prediction_days

    def get_data(self):
        data = yf.download(self.CryptoName, start=self.start_date, end=self.end_date)
        print(data)
        return data

    def create_windows(self, data):
        close_data = data['Close'].values
        x_data, y_data = [], []
        for i in range(self.prediction_days, len(close_data)):
            x_data.append(close_data[i-self.prediction_days:i])
            y_data.append(close_data[i])

        x_data, y_data = np.array(x_data), np.array(y_data)
        return x_data, y_data

    def train_lightgbm(self, x_train, y_train):
        model = LGBMRegressor()
        model.fit(x_train, y_train)
        return model

    def train_lstm(self, x_train, y_train):
        model = Sequential()
        
        model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
        model.add(LSTM(units=50))
        model.add(Dense(1))
        
        return model


    def forecast(self, model, x_test):
        predictions = model.predict(x_test)
        return predictions

    def evaluate_model(self, actual, predicted):
        mae = mean_absolute_error(actual, predicted)
        mse = mean_squared_error(actual, predicted)
        rmse = np.sqrt(mse)
        mape = np.mean(np.abs((actual - predicted) / actual)) * 100

        print(f"Mean Absolute Error (MAE): {mae}")
        print(f"Mean Squared Error (MSE): {mse}")
        print(f"Root Mean Squared Error (RMSE): {rmse}")
        print(f"Mean Absolute Percentage Error (MAPE): {mape}%")

        return mae, mse, rmse, mape

    def plot_forecast(self, real_data, predicted_data, future_dates):
        plt.figure(figsize=(14, 7))
        plt.plot(real_data.index, real_data['Close'], label='Preço Real')
        plt.plot(future_dates, predicted_data, label='Previsão', linestyle='--')
        plt.title(f'Previsão dos Preços de Fechamento da {self.CryptoName}')
        plt.xlabel('Data')
        plt.ylabel('Preço (USD)')
        plt.legend()
        plt.grid(True)
        plt.show()
        
    def forecast_future(self, model, last_data, days_to_predict):
        predicted_prices = []
        current_input = last_data.copy()

        for _ in range(days_to_predict):
            current_input_reshaped = current_input.reshape(1, -1)  # Formato que o modelo espera
            next_price = model.predict(current_input_reshaped)[0]  # Prever o próximo valor
            predicted_prices.append(next_price)
            
            # Atualizar a janela de entrada, removendo o valor mais antigo e adicionando a previsão
            current_input = np.append(current_input[1:], next_price)

        return predicted_prices

# Uso do modelo para prever o futuro
CryptoPredicterBoosting = CryptoPredicterWithBoosting('SOL-USD', '2020-01-01', '2024-09-01', 30)

sol_data = CryptoPredicterBoosting.get_data()

# Criando janelas de tempo
x_data, y_data = CryptoPredicterBoosting.create_windows(sol_data)

# Separando treino e teste
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, shuffle=False)

# Treinando o modelo LightGBM
lightgbm_model = CryptoPredicterBoosting.train_lightgbm(x_train, y_train)

# Treinando o modelo AdaBoost
adaboost_model = CryptoPredicterBoosting.train_adaboost(x_train, y_train)

# Prever os próximos 30 dias usando os últimos dados disponíveis como entrada
days_to_predict = 120
last_data = x_data[-1]  # Última janela de dados de treino

# Previsões com LightGBM
lightgbm_future_predictions = CryptoPredicterBoosting.forecast_future(lightgbm_model, last_data, days_to_predict)

# Previsões com AdaBoost
adaboost_future_predictions = CryptoPredicterBoosting.forecast_future(adaboost_model, last_data, days_to_predict)

# Gerar datas futuras para visualização
last_date = sol_data.index[-1]
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=days_to_predict, freq='D')

# Plotando previsões futuras
CryptoPredicterBoosting.plot_forecast(sol_data, lightgbm_future_predictions, future_dates)
CryptoPredicterBoosting.plot_forecast(sol_data, adaboost_future_predictions, future_dates)
```
